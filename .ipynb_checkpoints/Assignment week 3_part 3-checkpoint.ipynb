#### From part 1

```python
import wikipedia as wp
import pandas as pd 

#Import just the table from the site. I checked for the table name from the wiki page source.
table_url = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")

#determin the index of the table by first row
toronto_table = pd.read_html(table_url, header = [0])[0]
toronto_table.rename(columns={'Postal Code':'PostalCode'},inplace=True)

# Droping 'Not assigned' rows
toronto_table = toronto_table[toronto_table.Neighbourhood != 'Not assigned']

toronto_table
```

#### From part 2

```python
import requests, json, io

geo_url="http://cocl.us/Geospatial_data"
geo_csv = requests.get(geo_url).content
read_geo_csv = pd.read_csv(io.StringIO(geo_csv.decode('utf-8')))

# rename the first column to  merge dataframes based on Postalcode 
read_geo_csv.columns = ['PostalCode', 'Latitude', 'Longitude']
toronto_table = pd.merge(read_geo_csv, toronto_table, on='PostalCode')

# show the dataframe by order
toronto_table = toronto_table[['PostalCode', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']]

toronto_table
```

#### part 3 - Explore and cluster the neighborhoods in Toronto.

```python
from shapely.geometry import Point
import geopandas as gpd
import matplotlib.pyplot as plt

# load latitude and longitude into geopandas dataframe
toronto_table['Coordinates'] = list(zip(toronto_table.Longitude, toronto_table.Latitude))
toronto_table['Coordinates'] = toronto_table['Coordinates'].apply(Point)
geo_toronto_table = gpd.GeoDataFrame(toronto_table, geometry='Coordinates')

# set up map
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
ax = cities[cities.name == "Toronto"].plot(
    color='white', edgecolor='black')

# plot and show
geo_toronto_table.plot(ax=ax, color='red')
plt.show()
```

### Visualizing all the Neighbourhoods of the above data frame using Folium

```python
import folium 
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
import random
import pandas as pd
import numpy as np

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
for lat,lng,borough,neighbourhood in zip(toronto_table['Latitude'],toronto_table['Longitude'],toronto_table['Borough'],toronto_table['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
    
map_toronto
```

### Using KMeans clustering for the clsutering of the neighbourhoods

```python
k = 5
toronto_clustering = toronto_table.drop(['PostalCode','Borough','Neighbourhood','Coordinates'],1)

toronto_clustering
kmeans_table = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)

kmeans_table.labels_
```
